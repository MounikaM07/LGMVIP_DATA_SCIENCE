# **IMAGE TO PENCIL SKETCH**

```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
original_img = cv2.imread("/content/bird.jpg")
img = cv2.resize(original_img,(500,300))
cv2_imshow(img)
```

**CONVERTING INTO GRAYSCALE IMAGE**

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayscale_img)
```

**INVERTING THE GRAYSCALE IMAGE**

```python
inverted_img = cv2.bitwise_not(grayscale_img)
cv2_imshow(inverted_img)
```

```python
#applying Gaussian Function
inverted_blurry_img = cv2.GaussianBlur(inverted_img, (21,21),0)
cv2_imshow(inverted_blurry_img)
```

```python
inverted_blurred_img = 255- inverted_blurry_img
pencil_sketch = cv2.divide(grayscale_img , inverted_blurred_img , scale= 256)
```

**ORIGINAL IMAGE AND PENCIL SKETCH**

```python
cv2_imshow(img)
cv2_imshow(pencil_sketch)
```
